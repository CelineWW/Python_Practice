```python
# Create linechart function.
def linechart(x, y, legend, color):
    plt.plot(x,y)
    plt.xlabel('Date')
    plt.ylabel('Share Price')
    plt.title('Share Price')
    plt.legend(legend)
    plt.show()
```

```python
legend = ['open', 'high', 'low']
color = ['r', 'g', 'b']
linechart(df.index, df.iloc[:, 0:3], legend=legend, color=color)
```

```python
# Use legend to show which line represents which data. 
plt.figure(figsize=(10,5)) #set figure width and height.
legend = ['open', 'high', 'low']
plt.plot(df.index.values, df.iloc[:, 0:3].values)   # Here can x,y can use .values or without, get the same figure.
plt.xlabel('Date')
plt.ylabel('Share Price')
plt.legend(legend)  
plt.show()
```

```python
open=df.iloc[:,0:1].values # column 'Open'
high=df.iloc[:,1:2].values # column 'High'
low=df.iloc[:,2:3].values  # column 'Low'
```

```python
# Customize line color.
plt.figure(figsize=(10,5))
plt.plot(df.index.values, open, color='r', label='Open')
plt.plot(df.index.values, high, color='g', label='High')
plt.plot(df.index.values, low, color='b', label='Low')
plt.xlabel('Date')
plt.ylabel('Share Price')
plt.title('Share Price')

plt.legend()
plt.show()
```

```python
# Use marks to highlight data points.
plt.figure(figsize=(10,5))
plt.plot(df.index.values, open, 'r-*', label='Open') # straight line with datapoints.
plt.plot(df.index.values, high, 'g--o', label='High')  # dashline with datapoints.
plt.plot(df.index.values, low, 'bx', label='Low')   # No line, only datapoints.
plt.xlabel('Date')
plt.ylabel('Share Price')
plt.title('Share Price')

plt.legend()
plt.show()
```

## Use enumerate() for iterations.

```python
over1 = range(1,11)
print(over1)
print(*range(1,21,1))
runs1 = [i * 5 for i in over1]
print(runs1)
```

# Use DataTime dataset

```python
import datetime as DT
from matplotlib.dates import date2num

data = [(DT.datetime.strptime('2016-10-03', "%Y-%m-%d"), 772.559998),
        (DT.datetime.strptime('2016-10-04', "%Y-%m-%d"), 776.429993),
        (DT.datetime.strptime('2016-10-05', "%Y-%m-%d"), 776.469971),
        (DT.datetime.strptime('2016-10-06', "%Y-%m-%d"), 776.859985),
        (DT.datetime.strptime('2016-10-07', "%Y-%m-%d"), 775.080017 )]

data
```

```python
#Data.strftime(coverts the date to str)
x = [date.strftime("%Y-%m-%d") for (date, value) in data]
x
```

```python
y = [value for (date, value) in data] 
y
```

```python
x = [date2num(date) for (date, value) in data]
y = [value for (date, value) in data] 
fig = plt.figure()

# Plot the data as a red line with round markers
graph = fig.add_subplot(111)
graph.plot(x, y, 'r-o')
# Set the xtick locations
graph.set_xticks(x)
# Set the xtick labels
graph.set_xticklabels([date.strftime("%Y-%m-%d") for (date, value) in data])
plt.xlabel('Date')
plt.ylabel('Closing Value')
plt.title('Closing stock value of Alphabet Inc.') 
# Customize the grid
plt.grid(linestyle='-', linewidth='1', color='yellow')
plt.show()
```

# Bar Graph

```python
x = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
x_pos = np.arange(len(x))
print(x_pos)
```

```python
plt.bar(x, popularity, color='blue')
plt.xlabel("Languages")
plt.ylabel("Popularity")
plt.title("Popularity of Programming Language\n" + "Worldwide, Oct 2017 compared to a year ago")
plt.xticks(x_pos, x) #for x axis label
# Turn on the grid
plt.minorticks_on()
plt.grid(which='major', linestyle='-', linewidth='0.5', color='red')
# Customize the minor grid
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='black')
plt.show()
```

```python
# Add values on the top of the bar.
x = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
fig, ax = plt.subplots() #subplot will be used to create the fig and the axes
rects = ax.bar(x, popularity, color='b')
rects = ax.bar(x, popularity, color='b')
plt.xlabel("Languages")
plt.ylabel("Popularity")
plt.title("PopularitY of Programming Language\n" + "Worldwide, Oct 2017 compared to a year ago",bbox={'facecolor':'0.8', 'pad':5})

def autolabel(rects):
    """
    Attach text label above each bar displaying its height.
    """
    for rect in rects:
        height = rect.get_height()
        print(height, rect.get_x(), rect.get_width())
        ax.text(rect.get_x() + rect.get_width()/2., 
        height+0.25, '%.1f'%float(height),
        ha='center', va='bottom')  #ha=Horizontal Alignment va=vertical Aligbment
autolabel(rects)
plt.ylim(1,25) # change figure y limit to tolerate the maximun value shown up.
plt.show()          
```

```python
def myBar(x,y,xlable,ylabel,title):
    fig, ax = plt.subplots() 
    rects1 = ax.bar(x, y, color='b')
    plt.xlabel(xlable)
    plt.ylabel(ylabel)
    plt.title(title)
    autolabel(rects1)
    plt.ylim(1,25)
    plt.show()
```

```python
x = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
myBar(x,popularity,'Language','Popularity',"PopularitY of Programming Language\n" + 
      "Worldwide, Oct 2017 compared to a year ago")
```

# Pie Chart

```python
# Data to plot
languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
colors = ["Brown", "Blue", "yellow", "red", "green", "#8c564b"]
exp = (0.08, 0.05, 0, 0, 0, 0)
plt.pie(popularity,labels=languages,autopct='%1.1f%%', colors=colors, explode=exp, startangle=40)
plt.show()
```

```python
languages = 'Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++'
popuratity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
#colors = ['red', 'gold', 'yellowgreen', 'blue', 'lightcoral', 'lightskyblue']
colors = ["#1f77b4", "#ff7f0e", "#2ca02c", "#d62728", "#9467bd", "#8c564b"]
# explode 1st slice
explode = (0.1, 0, 0, 0, 0, 0)  
# Plot
plt.pie(popuratity, explode=explode, labels=languages, colors=colors,autopct='%1.1f%%', shadow=True, startangle=140)
plt.title("PopularitY of Programming Language\n" + "Worldwide, Oct 2017 compared to a year ago", bbox={'facecolor':'0.8', 'pad':5})
plt.show()
```
