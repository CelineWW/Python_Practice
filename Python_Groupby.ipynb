# Groupby: Split, Apply, Combine

```python
import pandas as pd
import numpy as np
```

```python
d={
    'team':["India","New zealand","Australia","India",
            "India","New zealand","Australia","India",
            "New zealand","Australia","New zealand","Australia"],
     'Run':[330,230,300,180,200,250,190,400,340,290,390,333],
     "Match":[2,4,2,1,5,6,3,2,5,3,1,1],
     "Year":['2012','2012','2012','2012','2013','2013','2013','2013','2014','2014','2014','2014']
  }
df=pd.DataFrame(d)
df
```

```python
df.info()
```

## Groupby Team

```python
df['team'].unique()
```

```python
g = df.groupby('team')
g
```

```python
g.head()
```

```python
for team, teamdata in g:
    print('team')
    print('-' * 40)
    print(teamdata)
    print('-' * 40)
```

## get_group('group name')

```python
g.get_group('Australia')
```

```python
g.get_group('India')
```

```python
g.get_group('New zealand')
```

```python
df.min() # min() in whole df in each column
```

```python
g.get_group('India').min() # min() in "India" group
```

```python
g.min() # min() in each group
```

```python
g.max()
```

```python
g.sum()
```

```python
g.mean()
```

## Groupby Year

```python
df['Year'].unique()
```

```python
g_yr = df.groupby('Year')
```

```python
g_yr.get_group('2013')
```

```python
g_yr.describe()
```

```python
g_yr['Run'].describe()
```
