# Concat DataFramef: Append by Row or Column

## <font color='green'>Append by Row</font>

```python
import pandas as pd
```

```python
state1={
      "date":['01-02-12','03-02-12','04-02-12','05-02-12'],
      "event":['hot','sunny','rainy','cold'],
      "wind-speed":[33,23,45,24],
      "temp":[20,21,15,18]
}
usstate1=pd.DataFrame(state1)
usstate1
```

```python
state2={
      "date":['01-02-12','03-02-12','04-02-12','05-02-12'],
      "event":['sunny','cold','cold','rainy'],
      "wind-speed":[12,10,9,14],
      "temp":[14,16,15,10]
}
usstate2=pd.DataFrame(state2)
usstate2
```

```python
df = pd.concat([usstate1, usstate2]) # index remains the same.
df
```

```python
df = pd.concat([usstate1, usstate2], ignore_index=True) # created a new index. orginal indices are ignored.
df
```

### Join dfs: keep the df name and its index.

```python
df = pd.concat([usstate1, usstate2], keys=['California', 'Texas'])
df
```

```python
df.loc['California']
```

```python
df.loc['Texas']
```

## <font color='green'>Append by Column</font> axis=1

```python
temp1={
      "date":['01-02-12','03-02-12','04-02-12','05-02-12'],
      "event":['sunny','cold','cold','rainy'],
      "temp":[14,16,15,10]
}
temp=pd.DataFrame(temp1)
temp
```

```python
ws={
      "date":['01-02-12','03-02-12','04-02-12','05-02-12'],
      "event":['sunny','cold','cold','rainy'],
      "wind-speed":[12,10,9,14],
}
wind_speed=pd.DataFrame(ws)
wind_speed
```

```python
df = pd.concat([temp, wind_speed], axis=1)  # add columns. all columns added.
df 
```

```python
df2 = pd.concat([temp, wind_speed]) # add rows.
df2
```

```python
df3 = pd.merge(temp, wind_speed) # merge combine both columns and rows.
df3
```
